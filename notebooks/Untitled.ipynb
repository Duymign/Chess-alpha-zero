```python
import os
import time
import sys
import subprocess
import logging

logging.basicConfig(
    level=logging.DEBUG,  # Hoặc INFO nếu bạn muốn ít hơn
    format="%(asctime)s [%(levelname)s] %(message)s",
    handlers=[logging.StreamHandler()]
)

play_data_dir = "data/play_data"
os.makedirs(play_data_dir, exist_ok=True)

data_files = os.listdir(play_data_dir)

def run_process(command, duration_minutes=20):
    import subprocess
    import time

    print(f"Đang chạy lệnh: {command}")
    start_time = time.time()
    end_time = start_time + (duration_minutes * 60)

    process = subprocess.Popen(
        command,
        shell=True,
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    try:
        while time.time() < end_time and process.poll() is None:
            time.sleep(5)

        if process.poll() is None:
            process.terminate()
            time.sleep(2)
            if process.poll() is None:
                process.kill()
            print(f"⏱ Đã dừng quá trình sau {duration_minutes} phút")

    except KeyboardInterrupt:
        process.terminate()
        print("⛔ Dừng bởi người dùng")

    # Đọc output
    stdout, stderr = process.communicate()

    # Hiển thị nếu có lỗi
    if process.returncode != 0:
        print("❌ Lỗi khi chạy lệnh:")
        print(stderr)

    return process.returncode


# Số lần lặp lại vòng huấn luyện
num_iterations = 1
self_play_time = 60
optimization_time = 60
# Thời gian cho mỗi giai đoạn (phút)

for i in range(num_iterations):
    print(f"\n=== Vòng huấn luyện {i+1}/{num_iterations} ===")
    

    # # Kiểm tra xem có dữ liệu đã được tạo ra không
    # play_files = [f for f in data_files if f.startswith("play_")]
    
    # if len(play_files) == 0:
    #     print("Không có dữ liệu play_data được tạo. Tiếp tục self-play...")
    #     print(run_process("python src/chess_zero/run.py --cmd self", self_play_time))
    
    # Chạy tối ưu hóa để huấn luyện mô hình
    print(f"Bắt đầu tối ưu hóa ")
    print(run_process("python src/chess_zero/run.py --cmd opt", optimization_time))
    
    # Kiểm tra mô hình đã được tạo ra chưa sau mỗi vòng lặp
    if os.path.exists("data/model/model_best_weight.h5"):
        print("Đã tìm thấy model_best_weight.h5!")
    else:
        print("Chưa tạo được model_best_weight.h5")

# Đánh giá mô hình cuối cùng
print("\n=== Đánh giá mô hình ===")
run_process("python src/chess_zero/run.py --cmd eval", 5)

# Hiển thị kết quả
print("\n=== Kết quả huấn luyện ===")
os.system("ls -la data/model/")
```
